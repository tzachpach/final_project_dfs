```python
salaries_df = pd.read_csv('nba_2020-2021_salaries.csv')
salaries_df = salaries_df[['First  Last', 'Date', 'H/A', 'GTime(ET)', 'Team pts', 'Opp pts', 
           'Start',  'GP', 'active', 'FDP', 'DKP', 'YHP', 'DoubleD', 'TripleD', 
           'FD Sal', 'FD Change', 'DK Sal', 'DK Change', 'YH Sal', 'YH Change', 
           'FD pos', 'DK pos', 'YH pos', 'ADI', 'VMI']]
```

```python
def convert_date(x):
    x = str(x)
    datetime_obj = pd.to_datetime(x, format='%Y%m%d')
    # Convert to desired format (YYYY-MM-DD)
    formatted_date = datetime_obj.strftime('%Y-%m-%d')
    return formatted_date

salaries_df['GAME_DATE'] = salaries_df['Date'].apply(lambda x: convert_date(x))
salaries_df['PLAYER_NAME'] = salaries_df['First  Last']
# convert_date(20210519.0)
```

```python
dk_pos = {5.0: 'C', 1.0: 'PG', 12.0: 'PG/SG', 23.0: 'SG/SF', 34.0: 'SF/PF', 45.0: 'PF/C', 2.0: 'SG', 4.0: 'PF', 3.0: 'SF', 13.0: 'PG/SF'}
yh_fd_pos = {2.0: 'SG', 3.0: 'SF', 4.0: 'PF', 5.0: 'C', 1.0: 'PG'}
salaries_df['DK pos'] = salaries_df['DK pos'].apply(lambda x: dk_pos[x] if not pd.isna(x) else x)
salaries_df['YH pos'] = salaries_df['YH pos'].apply(lambda x: yh_fd_pos[x] if not pd.isna(x) else x)
salaries_df['FD pos'] = salaries_df['FD pos'].apply(lambda x: yh_fd_pos[x] if not pd.isna(x) else x)
```

```python
df = pd.merge(games_df, salaries_df, on=['PLAYER_NAME', 'GAME_DATE'], how='left')
```
